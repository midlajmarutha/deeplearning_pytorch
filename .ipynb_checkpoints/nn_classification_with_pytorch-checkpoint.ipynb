# Building a classification model using PyTorch

```python
import sklearn
import torch
import torch.nn as nn
```

## Version 1

### Generating dataset

```python
from sklearn.datasets import make_circles

nsamples = 1000
X,y = make_circles(n_samples=nsamples,
                      noise = 0.03,
                      random_state=42)
```

```python
X[:5],y[:5]
```

```python
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=X[:,0],y=X[:,1],c=y,cmap=plt.cm.RdYlBu)
```

```python
X,y = torch.from_numpy(X).type(torch.float), torch.from_numpy(y).type(torch.float)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

### Building model

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
class CircleClassificationV0(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_layer1 = nn.Linear(in_features=2, out_features=5)
        self.linear_layer2 = nn.Linear(in_features=5, out_features=1)
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.linear_layer2(self.linear_layer1(x))
```

```python
model0 = CircleClassificationV0()
model0.to(device)
```

```python
# another way to build models
model0 = nn.Sequential(
    nn.Linear(in_features=2, out_features=5),
    nn.Linear(in_features=5, out_features=1))
model0
```

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)
model0.eval()
with torch.inference_mode():
    untrained_logits = model0(X_test.to(device))
untrained_logits.squeeze().shape, y_test.shape
```

```python
torch.round(untrained_logits[:5])
```

```python
pred_probs = torch.sigmoid(untrained_logits)
pred_probs[:5]
```

```python
pred = torch.round(pred_probs)
pred[:5], y_test[:5]
```

### Training and testing the model

```python
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(params = model0.parameters(),
                            lr = 0.01)
```

```python
def calculate_accuracy(y_true, y_pred):
    currect = torch.eq(y_true, y_pred).sum().item()
    accuracy = (currect / len(y_pred)) * 100
    return accuracy
```

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)
epochs_track = []
loss_track = []
X_train, X_test = X_train.to(device), X_test.to(device)
y_train, y_test = y_train.to(device), y_test.to(device)
epochs = 200
for epoch in range(epochs):
    epochs_track.append(epoch)
    model0.train()
    logits = model0(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(logits))
    loss = loss_fn(logits, y_train)
    loss_track.append(loss)
    accuracy = calculate_accuracy(y_true=y_train,
                                  y_pred=y_pred)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        model0.eval()
        with torch.inference_mode():
            test_pred = model0(X_test).squeeze()
        test_loss = loss_fn(test_pred, y_test)
        print(f"epochs: {epoch} | j_train: {loss:.5f} | j_test: {test_loss:.2f} | accuracy: {accuracy:.2f}%")
```

```python
model0.state_dict()
```

```python
plt.plot(epochs_track, torch.tensor(loss_track).numpy(),)
```

```python
# Looks like our model is not learning anything.
```

## Improved version

#### Methods to improve Deap learning models

* Increase number of hidden layers
* Increase number of hidden units
* Collect more data

* Tune learning rate
* Change activation functions


### Generating and splitting data

```python
from sklearn.datasets import make_circles
import torch
```

```python
X, y = make_circles(n_samples=1000,
                    noise=0.03,
                    random_state=42)
len(X), len(y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=X[:,0], y=X[:,1], c=y, cmap=plt.cm.RdYlBu)
```

```python
from sklearn.model_selection import train_test_split

X = torch.from_numpy(X).type(torch.float)
y = torch.from_numpy(y).type(torch.float)

X_train, X_test,y_train, y_test = train_test_split(X,y, train_size=0.8, random_state=42)
len(X_train), len(X_test), len(y_train), len(y_test)
```

### Building the model with non-linearity

```python
class CircleModelV2(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(in_features=2, out_features=10)
        self.layer2 = nn.Linear(in_features=10, out_features=10)
        self.layer3 = nn.Linear(in_features=10, out_features=1)
        self.relu = nn.ReLU()

    def forward(self, x):
        return self.layer3(self.relu(self.layer2(self.relu(self.layer1(x)))))
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
model2 = CircleModelV2().to(device)
model2
```

```python
X_train, X_test = X_train.to(device), X_test.to(device)
y_train, y_test = y_train.to(device), y_test.to(device)
```

### Training and testing the improved model

```python
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(params=model2.parameters(),
                            lr=0.1)
```

```python
torch.manual_seed(42)
torch.cuda.manual_seed(42)
epoch_count=[]
track_loss=[]
epochs = 2000
for epoch in range(epochs):
    model2.train()
    y_logits = model2(X_train).squeeze()
    y_pred = torch.round(torch.sigmoid(y_logits))
    epoch_count.append(epoch)
    loss = loss_fn(y_logits, y_train)
    track_loss.append(loss)
    train_acc = calculate_accuracy(y_train, y_pred)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        model2.eval()
        with torch.inference_mode():
            test_logits = model2(X_test).squeeze()
            test_pred = torch.round(torch.sigmoid(test_logits))
            test_loss = loss_fn(test_logits, y_test)
            test_acc = calculate_accuracy(y_test, test_pred)
        print(f"epochs: {epoch}, train_loss: {loss:.5f}, train_acc: {train_acc:.2f}% | test_loss: {test_loss:.5f}, test_acc: {test_acc:.2f}%")
```

```python
plt.plot(epoch_count, torch.tensor(track_loss).numpy())
```

## Yeeeey, 99.5% accuracy

```python
from pathlib import Path
MODEL_DIR = Path("model")
Path.mkdir(MODEL_DIR,
           parents=True,
           exist_ok=True)
MODEL_NAME = "binary_classification_improved.pth"
MODEL_PATH = MODEL_DIR / MODEL_NAME
torch.save(obj=model2.state_dict(),f=MODEL_PATH)
```
